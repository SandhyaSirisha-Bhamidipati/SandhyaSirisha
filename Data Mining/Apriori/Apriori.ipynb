### Task 2: Run apriory.py and Evaluate Results 

In this task, we have to find how we will be able to execute and print apriory results by making use only the apriory.py. In other words "WE DO NOT USE ANY OTHER LIBRARY FOR TASK 2!!!". 

We will have to execute apriori algorithm "3" times for different combinations of support and confidence.
Print the results of apriory for 'dataset.csv' by making use ONLY the provided methods. 

We add our reasoning (explain the result outcome) at the top of each case in a nice and readable way. 

```python
print ('# # # # # # # # #  Code for Task 2, Case:1 # # # # # # # # #') 


data_file = apriori.dataFromFile('dataset.csv')

items, rules = apriori.runApriori(data_file, 0.12, 0.40)
#print("Worst case")
print ('Case 1 Output:')
printResults(items, rules)



print ('Case 1 Reasoning: \n The rules should always have greater confidence and the value of confidence should be greater than the support value. Here, the support value is 0.12 (which is very low) and confidence is 0.4(low) so the no. of items and rules printed is very high. The confidence is 40% which means that if one element of the item set is already in the basket there is 40% chance that other item set is also added to the basket. Since, the support gives the frequency of the combination and confidence gives the probability of the occurrence., here for the combination of 0.12 and 0.4 it prints a huge list of items and rules.')

#print(items, rules)
```

```python
print ('# # # # # # # # #  Code for Task 2, Case:2 # # # # # # # # #') 


data_file = apriori.dataFromFile('dataset.csv')

items, rules = apriori.runApriori(data_file, 0.27, 0.50)
#print('Best case')
print ('Case 2 Output:')
printResults(items, rules)



print ('Case 2 (minimum support=0.27 and minimum confidence=0.5)')
print ('Case 2 Reasoning: \n The rules should always have greater confidence and the value of confidence should be greater than the support value. Here, the support value is 0.27 (which is average) and confidence is 0.5 (average) so the no. of items and rules printed is optimum (neither too huge nor too small). The confidence is 50% which means that if one element of the item set is already in the basket there is 50% chance that other item set is also added to the basket. Since, the support gives the frequency of the combination and confidence gives the probability of the occurrence., here for the combination of 0.27 and 0.50 it prints an acceptable (size wise) list of items and rules')




```

```python
print ('# # # # # # # # #  Code for Task 2, Case:3 # # # # # # # # #') 
data_file = apriori.dataFromFile('dataset.csv')

items, rules = apriori.runApriori(data_file, 0.2, 0.6)
#print('Average case')
print ('Case 3 Output:')
printResults(items, rules)
print ('Case 3 (minimum support=0.2 and minimum confidence=0.6)')
print ('Case 3 Reasoning: The rules should always have greater confidence and the value of confidence should be greater than the support value. Here, the support value is 0.2 (which is below-average ) and confidence is 0.6 (high) so the no. of items and rules printed is a more compared to case-2. The confidence is 60% which means that if one element of the item set is already in the basket there is 60% chance that other item set is also added to the basket. Since, the support gives the frequency of the combination and confidence gives the probability of the occurrence., here for the combination of 0.20 and 0.60 it prints an acceptable (size wise) list of items and rules')

```
