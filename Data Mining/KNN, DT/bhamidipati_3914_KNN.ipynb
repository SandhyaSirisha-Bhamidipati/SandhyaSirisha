```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import neighbors


#Extracting data
df_census = pd.read_csv('census1994.csv')
#Printing the dataset
print(df_census);
```

```python
#Printing the first five records of the dataset

first_five_recs = df_census.head();
print(first_five_recs);
```

```python
#Printing the first last records of the dataset

last_five_recs = df_census.tail();
print(last_five_recs);
```

```python
#Printing class names

for col in df_census.columns: 
    print(col)
```

```python
#Splitting the dataset 70% for training, and 30% for testing the classifier.

df_c = df_census[['Age', 'education-num', 'class']].copy()
df_c['class']=df_c['class'].map({' <=50K': 0, ' >50K': 1, ' <=50K.': 0, ' >50K.': 1})

X = df_c.iloc[:, 0:3];
print(X);
Y = df_c.iloc[:, 0];
print(Y);

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3);
print (X_train);
print (Y_test);
```

```python
#Create array from data values

array = df_c.values
X = array[:,0:16]
Y = array[:,2]

#select the features using chi square

test = SelectKBest(score_func=chi2, k=3)

#Fit the function for ranking the features by score

fit = test.fit(X_train, Y_train)

#Summarize scores numpy.set_printoptions(precision=3) print(fit.scores_)

#Apply the transformation on to dataset

features = fit.transform(X_train)
#print(features)

# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
# Fit the classifier to the data
knn.fit(X_train,Y_train)
```

```python
#show first 5 model predictions on the test data
knn.predict(X_test)[0:5]
```

```python
#check accuracy of our model on the test data
knn.score(X_test, Y_test)
```

```python
#create a new KNN model
knn_cv = KNeighborsClassifier(n_neighbors=3)
#train model with cv of 5 
cv_scores = cross_val_score(knn_cv, X, Y, cv=5)
#print each cv score (accuracy) and average them
print(cv_scores)
mean = np.mean(cv_scores)
print('mean:', mean)
#print(cv_score mean.format(np.mean(cv_scores)))
#print(cv_scores_mean.format(np.mean(cv_scores)))
```
