```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from sklearn import tree
%matplotlib inline

df_census = pd.read_csv('census1994.csv')
print(df_census);
df = df_census.replace('[?]',np.nan,regex=True)
df_census = df.dropna(how='any')
print(df_census.shape)
```

```python
first_five_recs = df_census.head();
print(first_five_recs);
```

```python
last_five_recs = df_census.tail();
print(last_five_recs);
```

```python
for col in df_census.columns: 
    print(col)
```

```python
#replacing string values with float values in dataframe to calculate euclidean distance:
df_c = df_census[['Age', 'education-num', 'class']].copy()
df_c['class']=df_c['class'].map({' <=50K': 0, ' >50K': 1, ' <=50K.': 0, ' >50K.': 1})

#Splitting the dataset 70% for training, and 30% for testing the classifier.

X = df_c.iloc[:, 0:3];
print(X);
Y = df_c.iloc[:, 0];
print(Y);

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3);
print (X_train);
print (Y_test);
```

```python
clf_gini=DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth= None, min_samples_leaf=1) 
clf_gini.fit(X_train, Y_train) 
```

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = None, min_samples_leaf = 1) 

clf_entropy.fit(X_train, Y_train) 
```

```python
y_pred = clf_gini.predict(X_test)
y_pred
```

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

```python
print(confusion_matrix(Y_test,y_pred));
```

```python
print(classification_report(Y_test,y_pred));
```

```python
print(confusion_matrix(Y_test,y_pred_en));
```

```python
print(classification_report(Y_test,y_pred_en)); 
```

```python
print ("Accuracy : ", 
    accuracy_score(Y_test,y_pred)*100)
```
